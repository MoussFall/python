```python
import tkinter as tk
from tkinter import messagebox

# Créer la fenêtre principale
window = tk.Tk()
window.title("Tic-Tac-Toe")

# Variables globales
current_player = "X"
board = [["" for _ in range(3)] for _ in range(3)]

# Fonction pour gérer le clic sur une case
def handle_click(row, col):
    global current_player
    if board[row][col] == "":
        board[row][col] = current_player
        buttons[row][col].config(text=current_player)
        
        # Vérifier s'il y a un gagnant
        if check_winner(current_player):
            messagebox.showinfo("Fin du jeu", f"Le joueur {current_player} a gagné !")
            reset_game()
        # Vérifier s'il y a un match nul
        elif is_board_full():
            messagebox.showinfo("Fin du jeu", "Match nul !")
            reset_game()
        else:
            # Changer de joueur
            current_player = "O" if current_player == "X" else "X"

# Fonction pour vérifier s'il y a un gagnant
def check_winner(player):
    for i in range(3):
        # Vérifier les lignes
        if board[i][0] == board[i][1] == board[i][2] == player:
            return True
        # Vérifier les colonnes
        if board[0][i] == board[1][i] == board[2][i] == player:
            return True
    # Vérifier les diagonales
    if board[0][0] == board[1][1] == board[2][2] == player:
        return True
    if board[0][2] == board[1][1] == board[2][0] == player:
        return True
    return False

# Fonction pour vérifier si la grille est pleine
def is_board_full():
    for row in board:
        if "" in row:
            return False
    return True

# Fonction pour réinitialiser le jeu
def reset_game():
    global current_player, board
    current_player = "X"
    board = [["" for _ in range(3)] for _ in range(3)]
    for i in range(3):
        for j in range(3):
            buttons[i][j].config(text="")

# Créer les boutons de la grille
buttons = []
for i in range(3):
    button_row = []
    for j in range(3):
        button = tk.Button(window, text="", width=10, height=5,
                           command=lambda row=i, col=j: handle_click(row, col))
        button.grid(row=i, column=j)
        button_row.append(button)
    buttons.append(button_row)

# Lancer la boucle principale
window.mainloop()
```
